```python
import os,shutil
import csv

def renameFile(srcPath):
    fname=srcPath.split('/')[2]
    return fname

def copyFile(srcPath,DestPath=None,fileName="entry"):#DestPath为目标文件夹路径，fileName为文件名
    DestPath='C:/Users/Matt/Desktop/Haps'
    if fileName=="entry":
        fileName=renameFile(srcPath)
    if not os.path.exists(DestPath):
        os.mkdir(DestPath)
    shutil.copyfile(srcPath,DestPath+'/'+fileName+'.hap')

def copyDir(srcPath,DestPath,fileName):
    DestPath='C:/HapModify/'
    if os.path.exists(DestPath+fileName+'MODIFY'):
        return
    shutil.copytree(srcPath,DestPath+fileName+'MODIFY')
    
```

```python
if __name__=='__main__':
    list=os.listdir('D:/hap')
    noPath=set()
    for dir in list:
        if dir=='.git':
            continue
        if os.path.exists('D:/hap/'+dir+'/build/outputs/hap/debug/phone/entry-debug-rich-unsigned.hap'):
            try:
                copyFile('D:/hap/'+dir+'/build/outputs/hap/debug/phone/entry-debug-rich-unsigned.hap',fileName=dir)
            except FileNotFoundError as e:
                print(e)
            # print(dir)
        elif os.path.exists('D:/hap/'+dir+'/entry/build/outputs/hap/debug/entry-debug-rich-unsigned.hap'):
            copyFile('D:/hap/'+dir+'/entry/build/outputs/hap/debug/entry-debug-rich-unsigned.hap',fileName=dir)
            # print(dir)
        else:
            noPath.add(dir)
    print(len(noPath))
```
